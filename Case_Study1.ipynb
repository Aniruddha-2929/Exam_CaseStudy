2.3 Remove Outliers

```python
#writing the function to remove all the outliers in
#all the columns and returning the column values having

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

for col in ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY']:
    df1 = remove_outliers_iqr(df1, col)
```

##6. Hypothesis testing

```python
# Imports (run once)
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as ws
from statsmodels.stats.proportion import proportions_ztest

# --- 6.1 Test 1: Income Difference (Two-sample t-test on log(income)) ---
# H0: mean(log_income | TARGET=1) == mean(log_income | TARGET=0)
income_default = np.log1p(df1.loc[df1['TARGET']==1, 'AMT_INCOME_TOTAL'].dropna())
income_nondef = np.log1p(df1.loc[df1['TARGET']==0, 'AMT_INCOME_TOTAL'].dropna())
tstat, pval = stats.ttest_ind(income_default, income_nondef, equal_var=False)   # Welch t-test
print('6.1 t-stat, p-value:', tstat, pval)

# --- 6.2 Test 2: Gender vs Default (Chi-square) ---
# H0: default rate independent of gender
ct = pd.crosstab(df1['CODE_GENDER'], df1['TARGET']).astype(int)   # rows: gender, cols: TARGET
chi2, pval, dof, exp = stats.chi2_contingency(ct)
print('6.2 chi2, p-value:', chi2, pval)

# --- 6.3 Test 3: Education Level & Default (Chi-square) ---
# H0: default rate independent of education level
ct_edu = pd.crosstab(df1['NAME_EDUCATION_TYPE'], df1['TARGET']).astype(int)
chi2_edu, pval_edu, dof_edu, exp_edu = stats.chi2_contingency(ct_edu)
print('6.3 chi2, p-value:', chi2_edu, pval_edu)

# (Optional ANOVA alternative: test mean EXT_SOURCE_1 across education groups)
# groups = [group.dropna() for _, group in df.groupby('NAME_EDUCATION_TYPE')['EXT_SOURCE_1']]
# fstat, p_anova = stats.f_oneway(*groups); print('ANOVA f, p:', fstat, p_anova)

# --- 6.4 Test 4: Past Refusals & Current Default (Two-proportion Z-test) ---
# H0: proportion(default | CNT_REFUSED>0) == proportion(default | CNT_REFUSED==0)
mask_refused = df1['CNT_REFUSED']>0
count = np.array([ df1.loc[mask_refused, 'TARGET'].sum(), df1.loc[~mask_refused, 'TARGET'].sum() ])
nobs  = np.array([ mask_refused.sum(), (~mask_refused).sum() ])
stat_z, pval_z = proportions_ztest(count, nobs)
print('6.4 z-stat, p-value:', stat_z, pval_z)

# --- 6.5 Test 5: One-sample proportion (company vs industry) ---
# Suppose industry_rate is given (e.g. 0.12). H0: company_default_rate == industry_rate
industry_rate = 0.12                     # <-- replace with actual industry rate
company_count = df1['TARGET'].sum()       # # defaults in company/sample
company_n = df1.shape[0]
z_stat_ind, pval_ind = proportions_ztest(company_count, company_n, value=industry_rate)
print('6.5 z-stat, p-value (vs industry_rate):', z_stat_ind, pval_ind)
```

##7. Insights and interpretation

```python
print("Income:", "Lower income linked to higher default" if pval<0.05 else "No income difference")
print("Gender:", "Default varies by gender" if pval<0.05 else "Gender has no impact")
print("Education:", "Default differs across education levels" if pval_edu<0.05 else "Education not a strong factor")
print("Past Refusals:", "Past refusals increase default risk" if pval_z<0.05 else "Refusals not strongly predictive")
print("Industry Comparison:", "Company default rate differs from industry" if pval_ind<0.05 else "Default rate matches industry")
```

```python
insights = {
    "Income": pval<0.05,
    "Gender": pval<0.05,
    "Education": pval_edu<0.05,
    "Past Refusals": pval_z<0.05,
    "Industry Default Rate": pval_ind<0.05
}
print({k: ("Significant — business impact present" if v else "Not significant") for k,v in insights.items()})
```
